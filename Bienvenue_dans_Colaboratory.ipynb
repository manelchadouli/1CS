```
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import nltk
import spacy
import re
from textblob import TextBlob
from textblob import Word
from nltk.corpus import stopwords
from nltk.corpus import state_union
import matplotlib.pyplot as plt
from pywaffle import Waffle
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from collections import Counter
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
import string
from nltk.stem import WordNetLemmatizer, PorterStemmer
from tokenizers import ByteLevelBPETokenizer
```

```
df = pd.read_csv('spooky.csv')
```

```
df.head(10)
```

```
# remove repetitive characters
rx = re.compile(r'([^\W\d_])\1{2,}')
for data in df['text']:
    data = re.sub(r'[^\W\d_]+',
                  lambda x: Word(rx.sub(r'\1\1', x.group())).correct() if rx.search(x.group())
                  else x.group(), data)
```

```
#homoglyphes manipulation
#we define function that map the homoglyphes to their right character and replace all the homoglyphes character with their standard form
def replace_homoglyphs(word):
    homoglyph_map = {
        '$' : 's',
        '0' : 'o',
        'ù' : 'u',
        'é' : 'e',
        'è' : 'e',
        'à' : 'a',
    }
    return ''.join([homoglyph_map.get(char, char) for char in word])


df['text'] = df['text'].replace(replace_homoglyphs(data))
```

```
# Transformer les entrées spéciales telles que les URL, les adresses e-mail et les balises HTML à une forme canonique.


# Assuming df is your DataFrame with a 'text' column
url_pattern = re.compile(r'(http://\S+)')
df['text'] = df['text'].replace(url_pattern, "URL")

html_pattern = re.compile(r'(<\S+>)')
df['text'] = df['text'].replace(html_pattern, "HTML")

email_pattern = re.compile(r'(\S+@\S+)')
df['text'] = df['text'].replace(email_pattern, "EMAIL")
```

```
#convert all to minuscule
df['text'] = df['text'].str.lower()
```

```
# remove pontuation
translator = str.maketrans('', '', string.punctuation + string.digits)
# df['text'] =df['text'].translate(translator)
df['text']= df['text'].apply(lambda x: x.translate(translator))
```

```
#remove stop words
nltk.download('stopwords')
en_stopwords = stopwords.words('english')
en_stopwords = set(en_stopwords)
for data in df['text']:
    words = data.split()
    words = [w for w in words if w not in en_stopwords]
    data = ' '.join(words)
```

```
df['text'] = df['text'].apply(lambda x: ' '.join([w for w in x.split() if w not in en_stopwords]))
```

```
#segmenter sur les espaces + ponctuations
import nltk
nltk.download('punkt')

token = []
for text in df['text']:
    tokens = nltk.word_tokenize(text)
    token.append(tokens)
```

```
words = []
for sentence in token:
    for word in sentence:
        words.append(word)
words
```

```
# 3- segmenter en sous-mots(bfe)
# Write the text data to a temporary file
from tempfile import NamedTemporaryFile
with NamedTemporaryFile(mode='w', delete=False) as temp_file:
    temp_file.write('\n'.join(df["text"].values.tolist()))

# Initialize the ByteLevelBPETokenizer
tokenizer = ByteLevelBPETokenizer()

# Train the tokenizer using the temporary file
tokenizer.train(files=temp_file.name, vocab_size=1000, min_frequency=10)

# Close and remove the temporary file
temp_file.close()

# Define a function to tokenize a string using the trained BPE tokenizer
def tokenize(text):
    encoded = tokenizer.encode(text)
    subwords = [tokenizer.id_to_token(id) for id in encoded.ids]
    return subwords

# Apply the tokenizer to each comment in the dataset
df["subword_comment"] = df["text"].apply(tokenize)

# Print the subword-tokenized comments
print(df["subword_comment"])
```

```
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```
# Pour chaque phrase représentait les entités nommées
#split comments into senctences by using "sent_tokenize" and then sentences to words using "word_tokenize", filter out stop words
#and non alphabitic words and uses ne_chunk to identify the named entities in each sentence (return tree structure)
sentences = []
for comment in df["text"]:
    sentences.extend(sent_tokenize(comment))
stop_words = set(stopwords.words('english'))
def preprocess(text):
    words = word_tokenize(text)
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words and word.isalpha()]
    tagged_words = pos_tag(filtered_words)
    named_entities = ne_chunk(tagged_words)
    return named_entities
for sentence in sentences:
    named_entities = preprocess(sentence)
    print(named_entities)
```

```
# use WordNetLemmatizer to lemmatize the words in each comment, and the PorterStemmer to stem the lemmatized words
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
def preprocess(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    # Filter out stop words and non-alphabetic words
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words and word.isalpha()]
    # Lemmatize the words
    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
    # Stem the words
    stemmed_words = [stemmer.stem(word) for word in lemmatized_words]
    # Return the preprocessed text as a string
    return " ".join(stemmed_words)
df["preprocessed_comment"] = df["text"].apply(preprocess)
print(df["preprocessed_comment"])
```

```
#En utilisant la fonction wordCloud, faire trois nuages de mots pour représenter les mots les plus utilisés par chaque auteur.

def generate_wordcloud(author):
    comments = df[df["author"] == author]["text"]
    text = " ".join(comments)
    # Créer le nuage de mots
    wordcloud = WordCloud(width=800, height=800, background_color="white").generate(text)
    # Afficher le nuage de mots
    plt.figure(figsize=(8, 8), facecolor=None)
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.title(f"most used words by {author}")
    plt.show()
```

```
#En utilisant la fonction wordCloud, montrer les 100 meilleurs mots positifs et négatifs utilisés par les auteurs
sid = SentimentIntensityAnalyzer()
def get_top_words(sentiment, n=100):
    comments = df["text"].tolist()
    word_counts = Counter()
    for comment in comments:
        # Supprimer la ponctuation
        comment = re.sub(r'[^\w\s]', '', comment)
        # Diviser en mots
        words = comment.split()
        # Pour chaque mot, calculer le score de sentiment et l'ajouter au compteur si le score est élevé
        for word in words:
            score = sid.polarity_scores(word)[sentiment]
            if score > 0.9:
                word_counts[word] += 1
    # Extraire les n mots les plus fréquents
    top_words = word_counts.most_common(n)
    return top_words

positive_words = get_top_words("pos")
negative_words = get_top_words("neg")
```

```
# Créer un nuage de mots pour les mots positifs
positive_wordcloud = WordCloud(width=800, height=800, background_color="white", colormap="Greens",
                               stopwords=STOPWORDS).generate_from_frequencies(dict(positive_words))

# Créer un nuage de mots pour les mots négatifs
negative_wordcloud = WordCloud(width=800, height=800, background_color="white", colormap="Reds",
                               stopwords=STOPWORDS).generate_from_frequencies(dict(negative_words))

# Afficher le nuage de mots pour les mots positifs
plt.figure(figsize=(10, 10))
plt.imshow(positive_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Mots positifs")
plt.show()
```

```
# Afficher le nuage de mots pour les mots négatifs
plt.figure(figsize=(10, 10))
plt.imshow(negative_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Mots négatifs")
plt.show()
```
